```python
import math

def calculate_entropy(data):
    data_size = len(data)
    unique_values = set(data)
    entropy = 0

    for value in unique_values:
        occurrence = data.count(value)
        probability = occurrence / data_size
        entropy -= probability * math.log2(probability)

    return entropy

# Load the dataset
dataset_file = "C:\\Users\This PC\Downloads\server-log.txt"  # Replace with the actual path to the dataset file
with open(dataset_file, 'r') as file:
    dataset = file.read()

# Extract relevant information from the dataset
entries = dataset.split('---\n')
attacks = []
for entry in entries:
    if 'Method' in entry:
        attack_info = {}
        for line in entry.split('\n'):
            if ':' in line:
                key, value = line.split(':')
                attack_info[key.strip()] = value.strip()
        attacks.append(attack_info)

# Extract relevant combinations of factors and calculate their entropy
combinations = ['Source IP', 'Target IP', 'Source Port', 'Target Port']
for attack in attacks:
    print(f"Attack Date/Time: {attack['Date/Time']}")
    for combination in combinations:
        factors = attack[combination].split('/')
        entropy = calculate_entropy(factors)
        print(f"{combination}: Entropy = {entropy}")
    print('---')
```
