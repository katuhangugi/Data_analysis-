```python
 #  Areas of Interest in kmÂ²
# AREAS_OF_INTEREST = {"SALZBURG": 7159, 
#                      "OBEROSTERREICH": 1595, 
#                      "KAERNTEN": 2581, 
#                      "TIROL": 3527, 
#                      "SUDTIROL": 1146, 
#                      "BAYERN": 760}

# You can add or remove from this list
UNUSED_PARAMETERS = ["FID", "FREQUENCY", "STD_VEL", "MAX_V_STDEV", "MIN_V_STDEV'", "MEAN_ACC", "STD_ACC", 
                     "MAX_ACC", "MIN_ACC'", "MEAN_A_STDEV", "STD_A_STDEV", "MEAN_COHERENCE", 
                     "STD_COHERENCE", "MIN_COHERENCE'","MAX_COHERENCE", "STD_STD_DEF", "MAX_STD_DEF", "MIN_STD_DEF'", 
                     "MEAN_SEASON_AMP", "STD_SEASON_AMP", "MAX_SEASON_AMP", "MIN_SEASON_AMP'", "MEAN_D20220713", "STD_D20220713"]


# Choose the AOI you're working on. Just change the name
# in the end, don't change other parameters!
# aoi_density_calculation = AREAS_OF_INTEREST["BAYERN"]

# Set the file name
# Use the track and the project for consistency
merged_file = 'EO4_T88'

# Paste the file directory
working_directory = r"C:\\Users\\This PC\\Desktop\\text_to_csv" + r"\\"

# Paste the output directory
output_directory = r"C:\\Users\\This PC\\Desktop\\text_to_csv\\"

# Necessary imports
import csv
import glob
import itertools
import os
import pandas as pd
from pathlib import Path

operating_directory = r"C:\\Users\\This PC\\Desktop\\text_to_csv"

files = glob.glob(working_directory + r'\*.txt')
file_names = [Path(i).stem for i in files]

for path in files:
    with open(path) as source:
        reader = csv.reader(source, delimiter=";")
        field_lines = []
        data_lines = []
        other_lines = []
        for line in reader:
            if line[0] == "FID":
                field_lines.append(line)
            elif line[0].startswith("Num"):
                other_lines.append(line)
            else:
                data_lines.append(line)

    fields = list(itertools.chain.from_iterable(field_lines))
    data = list(itertools.chain.from_iterable(data_lines))
    other = list(itertools.chain.from_iterable(other_lines))

    other_value_pair = {(split := element.rsplit(maxsplit=1))[0] : float(split[1]) for element in other}

    secondary_dataframe = pd.DataFrame(other_value_pair,
                                    index=[0])

    primary_dataframe = pd.DataFrame([data], columns=fields)
    primary_dataframe = (primary_dataframe.T.drop_duplicates().T).drop(columns=UNUSED_PARAMETERS, errors='ignore').replace('\.','', regex=True).replace(',','.', regex=True).astype(float)
    

    main_dataframe = pd.concat([primary_dataframe, secondary_dataframe],
                            axis=1)

    # main_dataframe['Area'] = aoi_density_calculation
    main_dataframe['% PS'] = round((main_dataframe['Num PS']/main_dataframe['Num MP'])*100, 2)
    main_dataframe['% DS'] = round((main_dataframe['Num DS']/main_dataframe['Num MP'])*100, 2)
    # main_dataframe['Density'] = round(main_dataframe['Num MP']/main_dataframe['Area'], 2)
    
    new_filename = Path(path).stem + '.xlsx'
    main_dataframe.to_excel(new_filename)

cwd = os.path.abspath('')
spreadsheets = os.listdir(cwd)

merger_dataframe = pd.DataFrame()

for file in spreadsheets:
    if file.endswith('.xlsx'):
        merger_dataframe = merger_dataframe.append(pd.read_excel(file),
                                                   ignore_index=True)

# create a list of duplicate file names
duplicated_file_names = []
for i in range(len(file_names)):
    duplicated_file_names.extend([file_names[i], file_names[i]])

# set the index of merger_dataframe to the duplicated file names
merger_dataframe.index = duplicated_file_names

# add the original file names as a column in merger_dataframe
merger_dataframe["SHAPEFILES"] = file_names * 2

merger_dataframe = merger_dataframe.drop("Unnamed: 0", axis=1).set_index("SHAPEFILES")

merger_dataframe.to_excel(merged_file + '.xlsx')


print("The program has ended")
```
