```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Load the data
data = pd.read_csv("C:\\Users\\This PC\\Downloads\\data_portfolio.csv")
# Calculate the mean returns and covariance matrix
returns = data.mean()
cov_matrix = data.cov()

# Define the objective function
def objective(weights):
    return -np.dot(weights, returns)

# Define the constraint functions
def constraint1(weights):
    return np.sum(weights) - 1.0

def constraint2(weights):
    return weights

def constraint3(weights):
    return np.dot(weights, returns) - target_return

# Set the target return (example value, you can change it)
target_return = 0.05

# Define the initial guess and bounds for the weights
num_assets = len(data.columns)
weights_guess = np.ones(num_assets) / num_assets
bounds = [(0, 1) for _ in range(num_assets)]

# Define the constraints
constraints = [{'type': 'eq', 'fun': constraint1},
               {'type': 'ineq', 'fun': constraint2},
               {'type': 'eq', 'fun': constraint3}]

# Solve the optimization problem
result = minimize(objective, weights_guess, method='SLSQP', bounds=bounds, constraints=constraints)

# Extract the optimal weights and the expected return of the portfolio
optimal_weights = result.x
expected_return = -result.fun

# Simulate portfolio returns
num_simulations = 10000
simulated_returns = np.random.multivariate_normal(returns, cov_matrix, num_simulations)

# Calculate the portfolio returns for each simulation
portfolio_returns = np.dot(simulated_returns, optimal_weights)

# Calculate the probability of target attainment
prob_target_attainment = np.mean(portfolio_returns >= target_return)

# Plot the probability of target attainment graph
plt.hist(portfolio_returns, bins=50, density=True, alpha=0.7)
plt.axvline(x=target_return, color='r', linestyle='--', label='Target Return')
plt.xlabel('Portfolio Returns')
plt.ylabel('Probability Density')
plt.title('Probability of Target Attainment')
plt.text(target_return+0.01, plt.gca().get_ylim()[1]*0.8, f'Probability = {prob_target_attainment:.2%}')
plt.legend()
plt.show()
```

```python
print("Optimal Weights:")
for asset, weight in zip(data.columns, optimal_weights):
    print(asset, ": ", weight)

print("Expected Return:", expected_return)
```
