```python
import pandas as pd

# Load the dataset
dataset_path = r'C:\Users\This PC\Downloads\MKTG6018_final_assessment\question_3.csv'
df = pd.read_csv(dataset_path)

# Display the first few rows of the dataset
print(df.head())
```

```python
# Check the data types and general information
print(df.info())

# Check basic statistical summary
print(df.describe())

# Check unique items in the "Item_Name" column
print(df['Item_Name'].unique())
```

```python
import matplotlib.pyplot as plt

# Calculate the total quantity ordered for each item
item_quantity = df.groupby('Item_Name')['Quantity'].sum()

# Sort the items in descending order of quantity
top_items = item_quantity.nlargest(5)

# Create a bar plot to visualize the most popular items
plt.figure(figsize=(10, 6))
top_items.plot(kind='bar')
plt.xlabel('Item Name')
plt.ylabel('Quantity Ordered')
plt.title('Top 5 Popular Food Items')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the total revenue generated by each item
item_revenue = df.groupby('Item_Name')['Total_products'].sum()

# Calculate the total cost of each item
item_cost = df.groupby('Item_Name')['Product_Price'].sum()

# Calculate the profitability of each item
item_profitability = item_revenue - item_cost

# Sort the items in descending order of profitability
top_profitable_items = item_profitability.nlargest(5)

# Create a bar plot to visualize the top profitable items
plt.figure(figsize=(10, 6))
top_profitable_items.plot(kind='bar')
plt.xlabel('Item Name')
plt.ylabel('Profitability')
plt.title('Top 5 Profitable Food Items')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the total number of orders for each customer
customer_order_count = df.groupby('Order_Number')['Item_Name'].count()

# Plot a histogram to visualize the distribution of order frequencies
plt.figure(figsize=(10, 6))
plt.hist(customer_order_count, bins=20)
plt.xlabel('Number of Orders')
plt.ylabel('Number of Customers')
plt.title('Distribution of Customer Order Frequencies')
plt.show()
```

```python
# Calculate the total order value for each customer
customer_order_value = df.groupby('Order_Number')['Total_products'].sum()

# Plot a histogram to visualize the distribution of order values
plt.figure(figsize=(10, 6))
plt.hist(customer_order_value, bins=20)
plt.xlabel('Order Value')
plt.ylabel('Number of Customers')
plt.title('Distribution of Customer Order Values')
plt.show()
```
