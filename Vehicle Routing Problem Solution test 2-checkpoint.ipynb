```python
from gurobipy import *

N = 8  # Number of customers (excluding depot)
K = 1  # Number of vehicles
bigM = 200
# Define data
c = 1  # Cost per unit distance
c_k = 25  # Capacity of each vehicle
e_i = [0, 912, 825, 65, 727, 15, 621, 170, 255]  # Earliest start time for each customer
l_i = [1236, 967, 870, 146, 782, 67, 702, 225, 324]  # Latest start time for each customer
P_max = 300  # Maximum time between earliest and latest start time
c_ei = 100  # Cost of service starting before earliest time
c_li = 50  # Cost of service starting after latest time
q_in = [0, 10, 30, 10, 10, 10, 20, 20, 20]  # Customer demand
d_ij = [
    [0, 5, 2, 3, 3, 3, 4, 4, 5],
    [5, 0, 3, 4, 2, 4, 2, 4, 3],
    [2, 3, 0, 3, 4, 3, 5, 3, 4],
    [3, 4, 3, 0, 4, 5, 3, 5, 2],
    [3, 2, 4, 4, 0, 2, 3, 2, 2],
    [3, 4, 3, 5, 2, 0, 3, 2, 3],
    [4, 2, 5, 3, 3, 3, 0, 2, 3],
    [4, 4, 3, 5, 2, 2, 2, 0, 3],
    [5, 3, 4, 2, 2, 3, 3, 3, 0]
]  # Travel distance between customers

# Create the model
model = Model()

# Define decision variables
x = {}  # Binary variables for determining the sequence of customer visits
for i in range(N + 1):
    for j in range(N + 1):
        x[i, j] = model.addVar(vtype=GRB.BINARY, name=f"x_{i}_{j}")

z = model.addVar(vtype=GRB.BINARY, name="z")  # Binary variable for vehicle activation

s = {}  # Service start time for each customer
for i in range(N + 1):
    s[i] = model.addVar(lb=e_i[i], ub=l_i[i] + P_max, name=f"s_{i}")

Delta_e = {}  # Variables to calculate early start time deviations
for i in range(N + 1):
    Delta_e[i] = model.addVar(vtype=GRB.CONTINUOUS, name=f"Delta_e_{i}")

Delta_l = {}  # Variables to calculate late start time deviations
for i in range(N + 1):
    Delta_l[i] = model.addVar(vtype=GRB.CONTINUOUS, name=f"Delta_l_{i}")

# Set objective function
model.setObjective(
    quicksum(c * d_ij[i][j] * x[i, j] for i in range(N + 1) for j in range(N + 1))
    + quicksum(c_k * z for k in range(K))
    + quicksum(c_ei * Delta_e[i] + c_li * Delta_l[i] for i in range(N + 1)),
    GRB.MINIMIZE,
)

# Add constraints
for k in range(K):
    model.addConstr(
        quicksum(x[i, 0] for i in range(N + 1)) == 1, name=f"constraint_1_{k}"
    )
    model.addConstr(
        quicksum(x[0, i] for i in range(N + 1)) == 1, name=f"constraint_2_{k}"
    )
    for j in range(N + 1):
        model.addConstr(
            quicksum(x[i, j] for i in range(N + 1)) == 1, name=f"constraint_3_{j}_{k}"
        )
        model.addConstr(
            quicksum(x[j, i] for i in range(N + 1)) == 1, name=f"constraint_4_{j}_{k}"
        )
    for m in range(N + 1):
        model.addConstr(
            quicksum(x[i, m] for i in range(N + 1))
            - quicksum(x[m, j] for j in range(N + 1))
            == 0,
            name=f"constraint_5_{m}_{k}",
        )
    for n in range(N + 1):
        model.addConstr(
            quicksum(q_in[i] * x[i, j] for i in range(N + 1) for j in range(N + 1))
            <= c_k * z,
            name=f"constraint_6_{n}_{k}",
        )

for i in range(1, N + 1):
    model.addConstr(s[i] >= e_i[i] - P_max, name=f"constraint_7_{i}")
    model.addConstr(s[i] <= l_i[i] + P_max, name=f"constraint_8_{i}")

for j in range(1, N + 1):
    model.addConstr(
        quicksum(
            x[i, j] * (s[i] + d_ij[i][j] + q_in[i]) for i in range(N + 1)
        )
        == s[j],
        name=f"constraint_9_{j}",
    )

for i in range(1, N + 1):
    model.addConstr(Delta_e[i] >= e_i[i] - s[i], name=f"constraint_10_{i}")
    model.addConstr(Delta_l[i] >= s[i] - l_i[i], name=f"constraint_11_{i}")

model.addConstr(s[0] == e_i[0], name="constraint_12")
model.addConstr(
    quicksum(
        quicksum(x[0, j] for j in range(N + 1)) for k in range(K)
    )
    == N,
    name="constraint_13",
)

for m in range(N + 1):
    model.addConstr(
        quicksum(x[i, m] for i in range(N + 1))
        - quicksum(x[m, j] for j in range(N + 1))
        == 1,
        name=f"constraint_14_{m}",
    )
    # Define the f_ij variables
f_ij = model.addVars(
    range(N + 1), range(N + 1), range(K), vtype=GRB.CONTINUOUS, name="f"
)

# ...

# Add constraint 15
for i in range(N + 1):
    for j in range(N + 1):
        for k in range(K):
            model.addConstr(
                f_ij[i, j, k] <= (N - 1) * x[i, j], name=f"constraint_15_{i}_{j}_{k}"
            )
model.addConstr(
    quicksum(
        quicksum(x[i, j] for i in range(N + 1)) for j in range(N + 1)
    )
    <= bigM * z,
    name="constraint_16",
)

# Optimize the model
model.optimize()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty graph
G = nx.Graph()

# Add nodes to the graph
for i in range(N + 1):
    G.add_node(i)

# Add edges to the graph based on the selected arcs
for i in range(N + 1):
    for j in range(N + 1):
        for k in range(K):
            if (i, j, k) in x and x[i, j, k].x > 0.5:  # Consider only arcs with a significant flow
                G.add_edge(i, j)

# Plot the graph
pos = nx.spring_layout(G)  # Layout algorithm for node positioning
nx.draw(G, pos, with_labels=True, node_color="skyblue", node_size=500)
labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Vehicle Routing Problem Solution")
plt.show()
```
