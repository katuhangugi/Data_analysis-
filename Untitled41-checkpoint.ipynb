```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

# Set the random seed for reproducibility
np.random.seed(0)
tf.random.set_seed(0)

# Define the parameters
eta = 0.5
y = 0.1
sigma = 0.3
r = 0.05
T = 1.0
N = 100
dt = T/N
t = np.linspace(0, T, N+1)

# Generate the Ornstein-Uhlenbeck process
Y0 = 0.0
Y = np.zeros(N+1)
Y[0] = Y0
for i in range(N):
    Y[i+1] = Y[i] + eta*(y-Y[i])*dt + sigma*np.sqrt(dt)*np.random.normal()

# Generate the optimal fraction of wealth for each value of the Ornstein-Uhlenbeck process
optimal_fraction = (eta*(y-r/eta+sigma**2/(2*eta)-Y))/(sigma**2)

# Define the neural network
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(1,), activation=None)
])

# Compile the model
model.compile(optimizer="sgd", loss="mse")

# Train the neural network
model.fit(Y.reshape(-1, 1), optimal_fraction.reshape(-1, 1), epochs=1000, verbose=0)

# Evaluate the neural network
optimal_fraction_nn = model.predict(Y.reshape(-1, 1)).flatten()

# Plot the results
import matplotlib.pyplot as plt
plt.plot(Y, optimal_fraction, label="Optimal")
plt.plot(Y, optimal_fraction_nn, label="Neural network")
plt.legend()
plt.xlabel("Y")
plt.ylabel("Optimal fraction")
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
S0 = 1
r = 0.05
sigma = 0.25
T = 1
N = 50
dt = T/N
x = 1

# Define the neural network architecture
def get_model():
    inputs = tf.keras.layers.Input(shape=(1,))
    x = tf.keras.layers.Dense(128, activation='relu')(inputs)
    x = tf.keras.layers.Dense(128, activation='relu')(x)
    x = tf.keras.layers.Dense(1, activation='sigmoid')(x)
    model = tf.keras.models.Model(inputs=inputs, outputs=x)
    return model

# Define the log utility function
def log_utility(c):
    return tf.math.log(c)

# Define the wealth equation
def wealth_equation(X, alpha, kappa, R):
    dX = alpha*X*R + (1-alpha)*X*r - kappa*X
    return dX

# Define the optimization problem
def objective_function(model_alpha, model_kappa, X):
    R = tf.random.normal((N,))
    wealth = [X]
    consumption = []
    alpha_values = []
    kappa_values = []

    for i in range(N):
        # Compute the trading strategy
        alpha = model_alpha(np.array([[wealth[-1]]]))
        alpha_values.append(alpha[0, 0])
        # Compute the propensity to consume
        kappa = model_kappa(np.array([[wealth[-1]]]))
        kappa_values.append(kappa[0, 0])
        # Compute the consumption rate
        c = kappa*wealth[-1]
        consumption.append(c)
        # Compute the wealth at the next time step
        dX = wealth_equation(wealth[-1], alpha, kappa, R[i])
        X_next = wealth[-1] + dX*dt
        wealth.append(X_next)

    # Compute the objective function value
    objective = log_utility(wealth[-1]) + tf.reduce_sum(log_utility(consumption))*dt
    return objective, wealth, alpha_values, kappa_values

# Define the theoretical optimal solution for kappa
def optimal_kappa(t):
    return 1/(T-t+1)

# Create the neural network models
model_alpha = get_model()
model_kappa = get_model()

# Define the optimizer
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

# Train the models
for i in range(5000):
    # Sample initial wealth
    X = np.random.uniform(0.5, 1.5)
    
    # Compute the objective function and its gradients
    with tf.GradientTape() as tape:
        objective, _, _, _ = objective_function(model_alpha, model_kappa, X)
    grads = tape.gradient(objective, model_alpha.trainable_weights + model_kappa.trainable_weights)
    
    # Update the model parameters
    optimizer.apply_gradients(zip(grads, model_alpha.trainable_weights + model_kappa.trainable_weights))
    
    # Print the objective function value every 100 iterations
    if i % 100 == 0:
        print(f"Iteration {i}: Objective function value = {objective.numpy():.4f}")

```
