```python
user_profile_visits = df.groupby('UserID')['ProfileID'].nunique()
print(user_profile_visits)
```

```python
top_active_users = user_profile_visits.nlargest(5)
print(top_active_users)
```

```python
user_country_counts = df['Country'].value_counts()
print(user_country_counts)
```

```python
import matplotlib.pyplot as plt

# Calculate the number of users from each country
user_country_counts = df['Country'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
user_country_counts.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of Users')
plt.title('User Distribution by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the total number of profiles visited by each user
user_profile_visits = df.groupby('UserID')['ProfileID'].nunique()

# Identify the top active users
top_active_users = user_profile_visits.nlargest(5)

# Create a bar plot
plt.figure(figsize=(10, 6))
top_active_users.plot(kind='bar')
plt.xlabel('User ID')
plt.ylabel('Number of Profiles Visited')
plt.title('Top Active Users')
plt.xticks(rotation=0)
plt.show()
```

```python
# Calculate the average visit duration for each user
user_avg_visit_duration = df.groupby('UserID')['Seconds'].mean()

# Identify users with the longest average visit duration
top_engaged_users = user_avg_visit_duration.nlargest(5)

# Create a bar plot
plt.figure(figsize=(10, 6))
top_engaged_users.plot(kind='bar')
plt.xlabel('User ID')
plt.ylabel('Average Visit Duration (Seconds)')
plt.title('Top Engaged Users')
plt.xticks(rotation=0)
plt.show()
```
