```python
import numpy as np
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam

# Step 1: Load the dataset
dataset_path = r"C:\Users\This PC\Downloads\gbpcny=x.csv"
df = pd.read_csv(dataset_path)

# Step 2: Add white Gaussian noise
std_dev = 0.1  # Set the standard deviation of the noise
num_noise_samples = len(df)  # Set the total number of noise samples
noise = np.random.normal(0, std_dev, num_noise_samples)
noisy_series = df['Close'].interpolate().values + noise

# Step 3: Apply ARIMA model to fit the noise-reduced series
model_arima = ARIMA(noisy_series, order=(1, 1, 1))
model_arima_fit = model_arima.fit()
arima_fitted = model_arima_fit.fittedvalues

# Step 4: Split the data into training and test sets
train_len = len(noisy_series) - 30
train_data = noisy_series[:train_len]
test_data = noisy_series[train_len:]

# Step 5: Prepare the data for LSTM training
def prepare_data(data, n_steps):
    X, y = [], []
    for i in range(len(data)-n_steps):
        X.append(data[i:i+n_steps])
        y.append(data[i+n_steps])
    return np.array(X), np.array(y)

n_steps = 1  # Number of previous days used for prediction
X_train, y_train = prepare_data(train_data, n_steps)
X_test, y_test = prepare_data(test_data, n_steps)

# Step 6: Train the LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(units=50, activation='relu', input_shape=(n_steps, 1)))
model_lstm.add(Dense(units=1))
model_lstm.compile(optimizer=Adam(learning_rate=0.005), loss='mean_squared_error')
model_lstm.fit(X_train, y_train, epochs=250, batch_size=1, verbose=1)

# Step 7: Make predictions with the LSTM model
y_pred_lstm = model_lstm.predict(X_test)

# Step 8: Align the lengths of original series and forecasts
forecast_arima_lstm = np.concatenate((arima_fitted, y_pred_lstm.flatten()))
forecast_arima_lstm = forecast_arima_lstm[-30:]  # Select the last 30 samples

# Step 9: Calculate the fit of the 30-day forecasts
fit_arima_lstm = mean_absolute_error(df['Close'].values[-30:], forecast_arima_lstm)

# Step 10: Print the fit of the 30-day forecasts
print("Fit of 30-Day Forecasts (ARIMA-LSTM):", fit_arima_lstm)
```

```python
import matplotlib.pyplot as plt

# Plotting the exchange rate and the 30-day forecasts
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Exchange Rate')
plt.plot(range(len(df)-30, len(df)), forecast_arima_lstm, label='30-Day Forecasts')
plt.title('Exchange Rate and 30-Day Forecasts')
plt.xlabel('Time')
plt.ylabel('Exchange Rate')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# Step 1: Load the dataset
dataset_path = r"C:\Users\This PC\Downloads\usdcny=x.csv"
df = pd.read_csv(dataset_path)

# Step 2: Add white Gaussian noise
std_dev = 0.1  # Set the standard deviation of the noise
num_noise_samples = len(df)  # Set the total number of noise samples
noise = np.random.normal(0, std_dev, num_noise_samples)
noisy_series = df['Close'].interpolate().values + noise

# Step 3: Apply ARIMA model to fit the noise-reduced series
model_arima = ARIMA(noisy_series, order=(1, 1, 1))
model_arima_fit = model_arima.fit()
arima_fitted = model_arima_fit.fittedvalues

# Step 4: Split the data into training and test sets
train_len = len(noisy_series) - 30
train_data = noisy_series[:train_len]
test_data = noisy_series[train_len:]

# Step 5: Prepare the data for LSTM training
def prepare_data(data, n_steps):
    X, y = [], []
    for i in range(len(data)-n_steps):
        X.append(data[i:i+n_steps])
        y.append(data[i+n_steps])
    return np.array(X), np.array(y)

n_steps = 1  # Number of previous days used for prediction
X_train, y_train = prepare_data(train_data, n_steps)
X_test, y_test = prepare_data(test_data, n_steps)

# Step 6: Train the LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(units=50, activation='relu', input_shape=(n_steps, 1)))
model_lstm.add(Dense(units=1))
model_lstm.compile(optimizer=Adam(learning_rate=0.005), loss='mean_squared_error')
history = model_lstm.fit(X_train, y_train, epochs=250, batch_size=1, verbose=1)

# Step 7: Make predictions with the LSTM model
y_pred_lstm = model_lstm.predict(X_test)

# Step 8: Align the lengths of original series and forecasts
forecast_arima_lstm = np.concatenate((arima_fitted, y_pred_lstm.flatten()))
forecast_arima_lstm = forecast_arima_lstm[-30:]  # Select the last 30 samples

# Step 9: Calculate the fit of the 30-day forecasts
fit_arima_lstm = mean_absolute_error(df['Close'].values[-30:], forecast_arima_lstm)

# Step 10: Print the fit of the 30-day forecasts
print("Fit of 30-Day Forecasts (ARIMA-LSTM):", fit_arima_lstm)

# Step 11: Plot the change in the optimal fitness value with the number of iterations
plt.plot(history.history['loss'])
plt.xlabel('Number of Iterations')
plt.ylabel('Optimal Fitness')
plt.title('Change in Optimal Fitness with Number of Iterations')
plt.show()
```
