```python
import gurobipy as gp
from gurobipy import GRB

def optimize_aircraft_assignment(flights, aircraft_types, fuel_cost, crew_cost, capacity, flight_time, max_flight_time):
    # Create a new model
    model = gp.Model("AircraftAssignment")

    # Create decision variables
    X = {}  # flight-aircraft assignment variables
    Y = {}  # aircraft selection variables

    for i in flights:
        for j in aircraft_types:
            X[i, j] = model.addVar(vtype=GRB.BINARY, name=f"X_{i}_{j}")

    for j in aircraft_types:
        Y[j] = model.addVar(vtype=GRB.BINARY, name=f"Y_{j}")

    # Set objective function
    objective = gp.quicksum(fuel_cost[i, j] * X[i, j] for i in flights for j in aircraft_types)
    objective += gp.quicksum(crew_cost[j] * Y[j] for j in aircraft_types)
    model.setObjective(objective, GRB.MINIMIZE)

    # Add constraints
    for i in flights:
        model.addConstr(gp.quicksum(X[i, j] for j in aircraft_types) == 1, f"FlightAssignment_{i}")

    for i in flights:
        for j in aircraft_types:
            model.addConstr(X[i, j] <= Y[j], f"AircraftSelection_{i}_{j}")

    for j in aircraft_types:
        model.addConstr(gp.quicksum(X[i, j] for i in flights) <= capacity[j] * Y[j], f"CapacityConstraint_{j}")

    for j in aircraft_types:
        model.addConstr(gp.quicksum(flight_time[i, j] * X[i, j] for i in flights) <= max_flight_time[j] * Y[j],
                        f"FlightTimeConstraint_{j}")

    # Optimize the model
    model.optimize()

    # Retrieve the results
    assignments = {}
    for i in flights:
        for j in aircraft_types:
            if X[i, j].x > 0.5:
                assignments[i] = j

    objective_value = model.objVal

    return assignments, objective_value

# Example usage
flights = [1, 2, 3, 4]
aircraft_types = [1, 2, 3]
fuel_cost = {(1, 1): 10, (1, 2): 15, (1, 3): 20,  # Example fuel cost data
             (2, 1): 12, (2, 2): 14, (2, 3): 18,
             (3, 1): 9, (3, 2): 16, (3, 3): 22,
             (4, 1): 11, (4, 2): 17, (4, 3): 19}
crew_cost = {1: 100, 2: 120, 3: 150}  # Example crew cost data
capacity = {1: 3, 2: 4, 3: 2}  # Example capacity data
flight_time = {(1, 1): 2, (1, 2): 3, (1, 3): 4,  # Example flight time data
               (2, 1): 3, (2, 2): 4, (2, 3): 5,
               (3, 1): 2, (3, 2): 4, (3, 3): 3,
               (4, 1): 3, (4, 2): 5, (4, 3): 4}
max_flight_time = {1: 8, 2: 10, 3: 6}  # Example maximum flight time data

assignments, objective_value = optimize_aircraft_assignment(flights, aircraft_types, fuel_cost, crew_cost, capacity,
                                                            flight_time, max_flight_time)

# Print the results
print("Flight - Aircraft Type Assignments:")
for flight, aircraft_type in assignments.items():
    print(f"Flight {flight}: Aircraft Type {aircraft_type}")

print("Objective Value:", objective_value)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_optimal_solution(assignments):
    # Create a directed graph
    G = nx.DiGraph()

    # Add nodes for flights and aircraft types
    flights = set(assignments.keys())
    aircraft_types = set(assignments.values())
    G.add_nodes_from(flights, bipartite=0)
    G.add_nodes_from(aircraft_types, bipartite=1)

    # Add edges for assignments
    for flight, aircraft_type in assignments.items():
        G.add_edge(flight, aircraft_type)

    # Specify positions of nodes for bipartite graph
    pos = nx.bipartite_layout(G, flights)

    # Draw nodes and edges
    nx.draw_networkx_nodes(G, pos, nodelist=flights, node_color='lightblue', node_size=500, alpha=0.8)
    nx.draw_networkx_nodes(G, pos, nodelist=aircraft_types, node_color='lightgreen', node_size=500, alpha=0.8)
    nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)

    # Add labels
    nx.draw_networkx_labels(G, pos)

    # Set plot options
    plt.axis('off')
    plt.title("Optimal Aircraft")
    plt.tight_layout()

    # Display the graph
    plt.show()

#  usage
assignments = {5: 10, 100: 500, 600: 600, 150: 600}  
draw_optimal_solution(assignments)
```
