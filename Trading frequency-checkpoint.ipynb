```python
#let's import the necessary libraries and load the data from the fileimport pandas as pd
import numpy as np

data = pd.read_excel("C:\\Users\\This PC\\Downloads\\investment decisions.xlsx", index_col=0)
```

```python
#We can check the first few rows of the data
print(data.head())
```

```python
returns = data.pct_change().dropna()
data
```

```python
#we can calculate the daily returns of each asset 
mu = returns.mean()
sigma = returns.std()

```

```python
#we can calculate the daily returns of each asset 
rf = 0.02 # risk-free rate
sharpe_ratio = (mu - rf) / sigma

```

```python
#We can then select the top 16 assets with the highest Sharpe ratio
selected_assets = sharpe_ratio.nlargest(16).index

```

```python
#we can calculate the expected return and volatility of a portfolio that invests equally in these 16 assets:
weights = np.ones(len(selected_assets)) / len(selected_assets)
portfolio_return = np.sum(mu[selected_assets] * weights)
portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(returns[selected_assets].cov(), weights)))
```

```python
#We can print the selected assets and the portfolio statistics:
print("Selected assets:", selected_assets)
print("Expected portfolio return:", portfolio_return)
print("Portfolio volatility:", portfolio_volatility)
```

```python
import matplotlib.pyplot as plt

# Plot the selected assets
plt.scatter(sigma[selected_assets], mu[selected_assets])

# Add labels for each point
for i, asset in enumerate(selected_assets):
    plt.annotate(asset, (sigma[asset], mu[asset]))

# Add the portfolio point
plt.scatter(portfolio_volatility, portfolio_return, color='red')

# Add labels and title
plt.xlabel('Volatility')
plt.ylabel('Expected Return')
plt.title('Efficient Frontier')

plt.show()
```
