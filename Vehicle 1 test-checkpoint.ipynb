```python
from gurobipy import *

N = 8  # Number of customers (excluding depot)
K = 3  # Number of vehicles

# Define data (Replace these with your actual data)
c = 1  # Cost per unit distance
c_ei = 100  # Cost of service starting before earliest time
c_li = 50   # Cost of service starting after latest time
P_max = 300  # Maximum time between earliest and latest start time

d_ij = [  # Travel distance between customers
    [0, 2, 2, 2, 3, 3, 3, 3, 3],
    [2, 0, 2, 2, 3, 3, 3, 3, 3],
    [2, 2, 0, 2, 3, 3, 3, 3, 3],
    [2, 2, 2, 0, 3, 3, 3, 3, 3],
    [3, 3, 3, 3, 0, 2, 2, 2, 2],
    [3, 3, 3, 3, 2, 0, 2, 2, 2],
    [3, 3, 3, 3, 2, 2, 0, 2, 2],
    [3, 3, 3, 3, 2, 2, 2, 0, 2],
    [3, 3, 3, 3, 2, 2, 2, 2, 0]
]
e_i = [0, 0, 10, 15, 20, 25, 30, 35, 40]  # Earliest start time for each customer
l_i = [0, 8, 12, 16, 20, 25, 30, 35, 40]  # Latest start time for each customer
q_in = [0, 5, 5, 5, 5, 5, 5, 5, 5]  # Customer demand
c_k = [20, 20, 20]  # Activation cost for each vehicle
M = 1e6
# Create the model
model = Model()

# Define decision variables
x = {}  # Binary variables for determining the sequence of customer visits
for i in range(N + 1):
    for j in range(N + 1):
        for k in range(K):
            x[i, j, k] = model.addVar(vtype=GRB.BINARY, name=f"x_{i}_{j}_{k}")

z = {}  # Binary variables for vehicle activation
for k in range(K):
    z[k] = model.addVar(vtype=GRB.BINARY, name=f"z_{k}")

s = {}  # Service start time for each customer
for i in range(N + 1):
    s[i] = model.addVar(lb=e_i[i], ub=l_i[i] + P_max, name=f"s_{i}")

y = {}  # Variable to prevent subtours
for i in range(N + 1):
    y[i] = model.addVar(vtype=GRB.CONTINUOUS, name=f"y_{i}")

Delta_e = {}  # Variables to calculate early start time deviations
for i in range(N + 1):
    Delta_e[i] = model.addVar(vtype=GRB.CONTINUOUS, name=f"Delta_e_{i}")

Delta_l = {}  # Variables to calculate late start time deviations
for i in range(N + 1):
    Delta_l[i] = model.addVar(vtype=GRB.CONTINUOUS, name=f"Delta_l_{i}")

# Set objective function
model.setObjective(
    quicksum(c * d_ij[i][j] * x[i, j, k] for i in range(N + 1) for j in range(N + 1) for k in range(K)) +
    quicksum(c_k[k] * z[k] for k in range(K)) +
    quicksum(c_ei * Delta_e[i] + c_li * Delta_l[i] for i in range(N + 1)),
    GRB.MINIMIZE
)

# Add constraints
model.addConstrs(quicksum(x[i, j, k] for j in range(N + 1) if j != i) == 1 for i in range(1, N + 1) for k in range(K))
model.addConstrs(quicksum(x[i, j, k] for i in range(N + 1) if i != j) == 1 for j in range(1, N + 1) for k in range(K))
model.addConstrs(quicksum(x[0, j, k] for j in range(1, N + 1)) == z[k] for k in range(K))
model.addConstrs(quicksum(x[i, j, k] * q_in[i] for i in range(N + 1)) <= 10 for j in range(N + 1) for k in range(K))
model.addConstrs(
    quicksum(x[i, j, k] for i in range(N + 1) if i != j) - quicksum(x[j, i, k] for i in range(N + 1) if i != j) == 0
    for j in range(1, N + 1) for k in range(K)
)
model.addConstrs(
    s[j] >= s[i] + q_in[i] + d_ij[i][j] - (1 - x[i, j, k]) * M
    for i in range(N + 1) for j in range(1, N + 1) for k in range(K)
)
model.addConstrs(y[i] >= s[i] - l_i[i] for i in range(N + 1))
model.addConstrs(y[i] >= e_i[i] - s[i] for i in range(N + 1))
model.addConstrs(Delta_e[i] >= e_i[i] - s[i] for i in range(N + 1))
model.addConstrs(Delta_l[i] >= s[i] - l_i[i] for i in range(N + 1))
model.addConstrs(Delta_e[i] <= P_max * (1 - x[0, i, k]) for i in range(1, N + 1) for k in range(K))
model.addConstrs(Delta_l[i] <= P_max * (1 - x[0, i, k]) for i in range(1, N + 1) for k in range(K))
model.addConstr(quicksum(z[k] for k in range(K)) >= 1)

# Optimize the model
model.optimize()

# Print the optimal solution
if model.status == GRB.OPTIMAL:
    print("Optimal solution found!")
    print("Objective value:", model.ObjVal)
    print("Vehicle routes:")
    for k in range(K):
        route = []
        for i in range(N + 1):
            for j in range(N + 1):
                if x[i, j, k].x > 0.5:
                    route.append(j)
        print(f"Vehicle {k+1}: {route}")
else:
    print("No solution found.")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph from the distance matrix
G = nx.Graph()
for i in range(N + 1):
    G.add_node(i)
for i in range(N + 1):
    for j in range(i + 1, N + 1):
        G.add_edge(i, j, weight=d_ij[i][j])

# Create a plot of the graph
pos = nx.spring_layout(G)
plt.figure(figsize=(8, 8))
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=500)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos, font_color='black', font_size=12)
plt.axis('off')
plt.title('Graph of Customers')
plt.show()

# Plot the optimal vehicle routes
plt.figure(figsize=(8, 8))
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=500)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos, font_color='black', font_size=12)
for k in range(K):
    route = []
    for i in range(N + 1):
        for j in range(N + 1):
            if x[i, j, k].x > 0.5:
                route.append(j)
    edges = [(route[i], route[i + 1]) for i in range(len(route) - 1)]
    nx.draw_networkx_edges(G, pos, edgelist=edges, width=2, edge_color='red')
plt.axis('off')
plt.title('Optimal Vehicle Routes')
plt.show()
```
