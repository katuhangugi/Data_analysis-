```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_rel

# Provided data (translated to English)
data = {
    'Front Side': {
        'Single Leg': {
            'Quadriceps': {'Jump Height': [0.216], 'RSI Mod': [0.17], 'Eccentric Energy': [0.0011], 'Concentric Energy': [0.0037], 'Eccentric RMS': [0.073], 'Concentric RMS': [0.188]},
            'Hamstrings': {'Jump Height': [0.216], 'RSI Mod': [0.17], 'Eccentric Energy': [0.0001], 'Concentric Energy': [0.0020], 'Eccentric RMS': [0.023], 'Concentric RMS': [0.094]},
            'Outer Gastrocnemius': {'Jump Height': [0.216], 'RSI Mod': [0.17], 'Eccentric Energy': [0.0003], 'Concentric Energy': [0.0016], 'Eccentric RMS': [0.059], 'Concentric RMS': [0.170]},
            'Anterior Tibialis': {'Jump Height': [0.216], 'RSI Mod': [0.17], 'Eccentric Energy': [0.0005], 'Concentric Energy': [0.0005], 'Eccentric RMS': [0.066], 'Concentric RMS': [0.089]},
            'Gluteus Maximus': {'Jump Height': [0.216], 'RSI Mod': [0.17], 'Eccentric Energy': [0.0003], 'Concentric Energy': [0.0003], 'Eccentric RMS': [0.042], 'Concentric RMS': [0.036]}
        },
        'Both Legs': {
            'Quadriceps': {'Jump Height': [0.512], 'RSI Mod': [0.63], 'Eccentric Energy': [0.0003], 'Concentric Energy': [0.0026], 'Eccentric RMS': [0.053], 'Concentric RMS': [0.191]},
            'Hamstrings': {'Jump Height': [0.512], 'RSI Mod': [0.63], 'Eccentric Energy': [0.0002], 'Concentric Energy': [0.0027], 'Eccentric RMS': [0.040], 'Concentric RMS': [0.159]},
            'Outer Gastrocnemius': {'Jump Height': [0.512], 'RSI Mod': [0.63], 'Eccentric Energy': [0.00007], 'Concentric Energy': [0.00039], 'Eccentric RMS': [0.019], 'Concentric RMS': [0.069]},
            'Anterior Tibialis': {'Jump Height': [0.512], 'RSI Mod': [0.63], 'Eccentric Energy': [0.0003], 'Concentric Energy': [0.0017], 'Eccentric RMS': [0.056], 'Concentric RMS': [0.112]},
            'Gluteus Maximus': {'Jump Height': [0.512], 'RSI Mod': [0.63], 'Eccentric Energy': [0.0007], 'Concentric Energy': [0.0019], 'Eccentric RMS': [0.0622], 'Concentric RMS': [0.1029]}
        }
    },
    'half_squat_6h_data': {
        'Single Leg': {
            'Quadriceps': {'Jump Height': [0.260], 'RSI Mod': [0.29], 'Eccentric Energy': [0.0021], 'Concentric Energy': [0.0043], 'Eccentric RMS': [0.246], 'Concentric RMS': [0.199]},
            'Hamstrings': {'Jump Height': [0.260], 'RSI Mod': [0.29], 'Eccentric Energy': [0.00006], 'Concentric Energy': [0.00021], 'Eccentric RMS': [0.023], 'Concentric RMS': [0.029]},
            'Outer Gastrocnemius': {'Jump Height': [0.260], 'RSI Mod': [0.29], 'Eccentric Energy': [0.0020], 'Concentric Energy': [0.0037], 'Eccentric RMS': [0.109], 'Concentric RMS': [0.245]},
            'Anterior Tibialis': {'Jump Height': [0.260], 'RSI Mod': [0.29], 'Eccentric Energy': [0.0025], 'Concentric Energy': [0.0012], 'Eccentric RMS': [0.143], 'Concentric RMS': [0.093]},
            'Gluteus Maximus': {'Jump Height': [0.260], 'RSI Mod': [0.29], 'Eccentric Energy': [0.00007], 'Concentric Energy': [0.00012], 'Eccentric RMS': [0.028], 'Concentric RMS': [0.035]}
        },
        'Both Legs': {
            'Quadriceps': {'Jump Height': [0.538], 'RSI Mod': [0.72], 'Eccentric Energy': [0.0001], 'Concentric Energy': [0.0013], 'Eccentric RMS': [0.052], 'Concentric RMS': [0.219]},
            'Hamstrings': {'Jump Height': [0.538], 'RSI Mod': [0.72], 'Eccentric Energy': [0.0009], 'Concentric Energy': [0.0066], 'Eccentric RMS': [0.068], 'Concentric RMS': [0.204]},
            'Outer Gastrocnemius': {'Jump Height': [0.538], 'RSI Mod': [0.72], 'Eccentric Energy': [0.00002], 'Concentric Energy': [0.0023], 'Eccentric RMS': [0.014], 'Concentric RMS': [0.141]},
            'Anterior Tibialis': {'Jump Height': [0.538], 'RSI Mod': [0.72], 'Eccentric Energy': [0.0011], 'Concentric Energy': [0.0016], 'Eccentric RMS': [0.106], 'Concentric RMS': [0.165]},
            'Gluteus Maximus': {'Jump Height': [0.538], 'RSI Mod': [0.72], 'Eccentric Energy': [0.00028], 'Concentric Energy': [0.00030], 'Eccentric RMS': [0.042], 'Concentric RMS': [0.050]}
        }
    },
    'half_squat_24h_data': {
        'Single Leg': {
            'Quadriceps': {'Jump Height': [0.267], 'RSI Mod': [0.30], 'Eccentric Energy': [0.0089], 'Concentric Energy': [0.0188], 'Eccentric RMS': [0.149], 'Concentric RMS': [0.321]},
            'Hamstrings': {'Jump Height': [0.267], 'RSI Mod': [0.30], 'Eccentric Energy': [0.0007], 'Concentric Energy': [0.0041], 'Eccentric RMS': [0.053], 'Concentric RMS': [0.160]},
            'Outer Gastrocnemius': {'Jump Height': [0.267], 'RSI Mod': [0.30], 'Eccentric Energy': [0.0001], 'Concentric Energy': [0.0009], 'Eccentric RMS': [0.024], 'Concentric RMS': [0.090]},
            'Anterior Tibialis': {'Jump Height': [0.267], 'RSI Mod': [0.30], 'Eccentric Energy': [0.0021], 'Concentric Energy': [0.0065], 'Eccentric RMS': [0.131], 'Concentric RMS': [0.189]},
            'Gluteus Maximus': {'Jump Height': [0.267], 'RSI Mod': [0.30], 'Eccentric Energy': [0.0038], 'Concentric Energy': [0.0009], 'Eccentric RMS': [0.143], 'Concentric RMS': [0.061]}
        },
        'Both Legs': {
            'Quadriceps': {'Jump Height': [0.560], 'RSI Mod': [0.73], 'Eccentric Energy': [0.0020], 'Concentric Energy': [0.0061], 'Eccentric RMS': [0.121], 'Concentric RMS': [0.265]},
            'Hamstrings': {'Jump Height': [0.560], 'RSI Mod': [0.73], 'Eccentric Energy': [0.0026], 'Concentric Energy': [0.0054], 'Eccentric RMS': [0.124], 'Concentric RMS': [0.165]},
            'Outer Gastrocnemius': {'Jump Height': [0.560], 'RSI Mod': [0.73], 'Eccentric Energy': [0.0003], 'Concentric Energy': [0.0023], 'Eccentric RMS': [0.075], 'Concentric RMS': [0.137]},
            'Anterior Tibialis': {'Jump Height': [0.560], 'RSI Mod': [0.73], 'Eccentric Energy': [0.0011], 'Concentric Energy': [0.0012], 'Eccentric RMS': [0.085], 'Concentric RMS': [0.120]},
            'Gluteus Maximus': {'Jump Height': [0.560], 'RSI Mod': [0.73], 'Eccentric Energy': [0.0030], 'Concentric Energy': [0.0008], 'Eccentric RMS': [0.142], 'Concentric RMS': [0.084]}
        }
    },
    'bulgarian_split_squat_6h_data': {
        'Single Leg': {
            'Quadriceps': {'Jump Height': [0.276], 'RSI Mod': [0.33], 'Eccentric Energy': [0.0070], 'Concentric Energy': [0.0122], 'Eccentric RMS': [0.187], 'Concentric RMS': [0.276]},
            'Hamstrings': {'Jump Height': [0.276], 'RSI Mod': [0.33], 'Eccentric Energy': [0.0014], 'Concentric Energy': [0.0024], 'Eccentric RMS': [0.080], 'Concentric RMS': [0.116]},
            'Outer Gastrocnemius': {'Jump Height': [0.276], 'RSI Mod': [0.33], 'Eccentric Energy': [0.00003], 'Concentric Energy': [0.0008], 'Eccentric RMS': [0.018], 'Concentric RMS': [0.078]},
            'Anterior Tibialis': {'Jump Height': [0.276], 'RSI Mod': [0.33], 'Eccentric Energy': [0.0034], 'Concentric Energy': [0.0025], 'Eccentric RMS': [0.183], 'Concentric RMS': [0.147]},
            'Gluteus Maximus': {'Jump Height': [0.276], 'RSI Mod': [0.33], 'Eccentric Energy': [0.0005], 'Concentric Energy': [0.0004], 'Eccentric RMS': [0.072], 'Concentric RMS': [0.068]}
        },
        'Both Legs': {
            'Quadriceps': {'Jump Height': [0.568], 'RSI Mod': [0.72], 'Eccentric Energy': [0.0031], 'Concentric Energy': [0.0132], 'Eccentric RMS': [0.113], 'Concentric RMS': [0.259]},
            'Hamstrings': {'Jump Height': [0.568], 'RSI Mod': [0.72], 'Eccentric Energy': [0.0001], 'Concentric Energy': [0.0008], 'Eccentric RMS': [0.016], 'Concentric RMS': [0.070]},
            'Outer Gastrocnemius': {'Jump Height': [0.568], 'RSI Mod': [0.72], 'Eccentric Energy': [0.0003], 'Concentric Energy': [0.0454], 'Eccentric RMS': [0.047], 'Concentric RMS': [0.339]},
            'Anterior Tibialis': {'Jump Height': [0.568], 'RSI Mod': [0.72], 'Eccentric Energy': [0.0004], 'Concentric Energy': [0.0024], 'Eccentric RMS': [0.075], 'Concentric RMS': [0.145]},
            'Gluteus Maximus': {'Jump Height': [0.568], 'RSI Mod': [0.72], 'Eccentric Energy': [0.0011], 'Concentric Energy': [0.0021], 'Eccentric RMS': [0.082], 'Concentric RMS': [0.120]}
        }
    },
    'bulgarian_split_squat_24h_data': {
        'Single Leg': {
            'Quadriceps': {'Jump Height': [0.286], 'RSI Mod': [0.30], 'Eccentric Energy': [0.0004], 'Concentric Energy': [0.0008], 'Eccentric RMS': [0.042], 'Concentric RMS': [0.080]},
            'Hamstrings': {'Jump Height': [0.286], 'RSI Mod': [0.30], 'Eccentric Energy': [0.0011], 'Concentric Energy': [0.0053], 'Eccentric RMS': [0.069], 'Concentric RMS': [0.156]},
            'Outer Gastrocnemius': {'Jump Height': [0.286], 'RSI Mod': [0.30], 'Eccentric Energy': [0.00001], 'Concentric Energy': [0.0045], 'Eccentric RMS': [0.009], 'Concentric RMS': [0.134]},
            'Anterior Tibialis': {'Jump Height': [0.286], 'RSI Mod': [0.30], 'Eccentric Energy': [0.0005], 'Concentric Energy': [0.0010], 'Eccentric RMS': [0.079], 'Concentric RMS': [0.107]},
            'Gluteus Maximus': {'Jump Height': [0.286], 'RSI Mod': [0.30], 'Eccentric Energy': [0.0010], 'Concentric Energy': [0.0013], 'Eccentric RMS': [0.079], 'Concentric RMS': [0.089]}
        },
        'Both Legs': {
            'Quadriceps': {'Jump Height': [0.547], 'RSI Mod': [0.75], 'Eccentric Energy': [0.0002], 'Concentric Energy': [0.0047], 'Eccentric RMS': [0.042], 'Concentric RMS': [0.157]},
            'Hamstrings': {'Jump Height': [0.547], 'RSI Mod': [0.75], 'Eccentric Energy': [0.0003], 'Concentric Energy': [0.0023], 'Eccentric RMS': [0.074], 'Concentric RMS': [0.137]},
            'Outer Gastrocnemius': {'Jump Height': [0.547], 'RSI Mod': [0.75], 'Eccentric Energy': [0.00002], 'Concentric Energy': [0.0119], 'Eccentric RMS': [0.009], 'Concentric RMS': [0.274]},
            'Anterior Tibialis': {'Jump Height': [0.547], 'RSI Mod': [0.75], 'Eccentric Energy': [0.0019], 'Concentric Energy': [0.0013], 'Eccentric RMS': [0.129], 'Concentric RMS': [0.139]},
            'Gluteus Maximus': {'Jump Height': [0.547], 'RSI Mod': [0.75], 'Eccentric Energy': [0.0015], 'Concentric Energy': [0.0082], 'Eccentric RMS': [0.088], 'Concentric RMS': [0.251]}
        }
    }
}


# Convert the data into a Pandas DataFrame
dfs = []  # Initialize an empty list to store DataFrames

for exercise, legs_data in data.items():
    for leg_type, muscles_data in legs_data.items():
        for muscle, parameters in muscles_data.items():
            df = pd.DataFrame(parameters)
            df['Exercise'] = exercise
            df['Leg Type'] = leg_type
            df['Muscle'] = muscle
            dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
df_combined = pd.concat(dfs, ignore_index=True)

# Drop rows with missing values
df_combined = df_combined.dropna()

# Perform paired t-tests for selected parameters (example: Jump Height)
ttest_results = {}
parameters_of_interest = ['Jump Height', 'RSI Mod', 'Eccentric Energy', 'Concentric Energy', 'Eccentric RMS', 'Concentric RMS']

for parameter in parameters_of_interest:
    front_side_data = df_combined[(df_combined['Exercise'] == 'Front Side') & (df_combined['Leg Type'] == 'Single Leg')][parameter]
    both_legs_data = df_combined[(df_combined['Exercise'] == 'Both Legs') & (df_combined['Leg Type'] == 'Single Leg')][parameter]

    # Check if lengths match
    if len(front_side_data) == len(both_legs_data):
        ttest_results[parameter] = ttest_rel(front_side_data, both_legs_data)
    else:
        print(f"Skipping {parameter} due to unequal lengths")

# Print t-test results
for parameter, result in ttest_results.items():
    print(f"T-test for {parameter}: p-value = {result.pvalue:.4f}")

# Plotting example: Jump Height
plt.figure(figsize=(10, 6))
sns.boxplot(x='Muscle', y='Jump Height', hue='Leg Type', data=df_combined, palette='Set2')
plt.title('Comparison of Jump Height for Different Muscles and Leg Types')
plt.show()
```
