```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Define the portfolio_volatility() function
def portfolio_volatility(weights, covariances):
    return np.sqrt(np.dot(weights.T, np.dot(covariances, weights)))

# Load the data from the CSV file
data = pd.read_excel("C:\\Users\\This PC\\Downloads\\investment decisions.xlsx", index_col=0)


# Randomly select 16 assets
assets = np.random.choice(list(data.columns), 16, replace=False)

# Extract the total return data for the selected assets
selected_data = data[assets]

# Calculate the expected returns and covariances
returns = selected_data.mean()
covariances = selected_data.cov()

# Calculate the risk-free rate
rf_rate = 0.02

# Create a range of portfolio weights
num_ports = 1000
weights = np.zeros((num_ports, len(selected_data.columns)))
for i in range(num_ports):
    weight = np.random.random(len(selected_data.columns))
    weight /= np.sum(weight)
    weights[i] = weight

# Calculate the portfolio returns and volatilities
port_returns = np.dot(weights, returns)
port_volatilities = np.zeros(num_ports)
for i in range(num_ports):
    port_volatilities[i] = portfolio_volatility(weights[i], covariances)

# Calculate the efficient frontier
max_returns = np.max(port_returns)
max_volatility = np.max(port_volatilities)
min_volatility = np.min(port_volatilities)
frontier_returns = np.linspace(0, max_returns, 100)
frontier_volatility = np.zeros_like(frontier_returns)
x0 = np.ones(len(selected_data.columns)) / len(selected_data.columns)
for i in range(len(frontier_returns)):
    cons = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1},
            {'type': 'eq', 'fun': lambda x: np.dot(x.T, returns) - frontier_returns[i]})
    res = minimize(portfolio_volatility, x0, args=(covariances,), method='SLSQP', constraints=cons)
    frontier_volatility[i] = res.fun

# Plot the efficient frontier and the capital market line
plt.scatter(port_volatilities, port_returns)
plt.xlabel('Volatility')
plt.ylabel('Expected Return')
plt.title('Portfolio volitilities')
plt.plot(frontier_volatility, frontier_returns, 'r--', label='Efficient Frontier')
plt.plot([0, max_volatility], [rf_rate, max_returns], 'g-', label='Capital Market Line')
plt.legend()
plt.show()
```
