```python
import numpy as np
import pandas as pd
import pyhht
from statsmodels.tsa.arima.model import ARIMA
import pywt
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

# Step 1: Load the dataset
dataset_path = r"C:\Users\This PC\Downloads\gbpcny=x.csv"
df = pd.read_csv(dataset_path)

# Step 2: Add white Gaussian noise
std_dev = 0.2  # Set the standard deviation of the noise
num_noise_samples = len(df)  # Set the total number of noise samples
noise = np.random.normal(0, std_dev, num_noise_samples)
noisy_series = df['Close'].interpolate().values + noise

# Step 3: Apply CEEMDAN-ML decomposition
decomposer = pyhht.EMD(noisy_series)
imfs = decomposer.decompose()

# Step 4: Calculate Pearson correlation coefficient
correlations = []
num_imfs = len(imfs)
for i in range(num_imfs):
    correlation = np.corrcoef(df['Close'].values, imfs[i])[0, 1]
    correlations.append(correlation)

# Step 5: Apply wavelet packet denoising
threshold = 0.2  # Set the correlation threshold
denoised_imfs = []
for i in range(num_imfs):
    if correlations[i] < threshold:
        coeffs = pywt.wavedec(imfs[i], wavelet='db4', level=5)
        coeffs[1:] = (pywt.threshold(c, std_dev*2, mode='soft') for c in coeffs[1:])
        denoised_imf = pywt.waverec(coeffs, wavelet='db4')
        denoised_imfs.append(denoised_imf)
    else:
        denoised_imfs.append(imfs[i])

# Step 6: Reconstruct the final denoised series
denoised_series = np.sum(denoised_imfs, axis=0)

# Step 7: Compare with ARIMA fitting result
model = ARIMA(noisy_series, order=(1, 1, 1))
model_fit = model.fit()
arima_fitted = model_fit.fittedvalues

# Step 8: Calculate comparison indicators
denoised_series = pd.Series(denoised_series).fillna(method='ffill').values
valid_indices = np.isfinite(df['Close'].values) & np.isfinite(denoised_series)
mae = mean_absolute_error(df['Close'].values[valid_indices], denoised_series[valid_indices])
mse = mean_squared_error(df['Close'].values[valid_indices], denoised_series[valid_indices])
mape = mean_absolute_percentage_error(df['Close'].values[valid_indices], denoised_series[valid_indices])
rmse = np.sqrt(mse)

# Step 9: Model forecasting and evaluation (repeat 10 times and average the results)
num_predictions = 30  # Set the number of predictions
num_iterations = 10  # Set the number of iterations

def arima_forecast(series, num_predictions):
    model = ARIMA(series, order=(1, 1, 1))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=num_predictions)
    return forecast

predictions = []
for _ in range(num_iterations):
    forecasts = arima_forecast(noisy_series, num_predictions)
    predictions.append(forecasts)

# Take the average of the predictions
average_predictions = np.mean(predictions, axis=0)

# Calculate the fitting degree of the ARIMA model
arima_fit_degree = model_fit.aic

# Plotting
plt.figure(figsize=(12, 6))

# Original series
plt.subplot(2, 2, 1)
plt.plot(df['Close'], color='blue')
plt.title('Original Series')
plt.xlabel('Time')
plt.ylabel('Exchange Rate')

# Noisy series
plt.subplot(2, 2, 2)
plt.plot(noisy_series, color='red')
plt.title('Noisy Series')
plt.xlabel('Time')
plt.ylabel('Exchange Rate')

# Denoised series
plt.subplot(2, 2, 3)
plt.plot(df.index, denoised_series, color='green')
plt.title('Denoised Series')
plt.xlabel('Time')
plt.ylabel('Exchange Rate')

# ARIMA fitted values
plt.subplot(2, 2, 4)
plt.plot(df.index, arima_fitted, color='purple')
plt.title('ARIMA Fitted Values')
plt.xlabel('Time')
plt.ylabel('Exchange Rate')

plt.tight_layout()
plt.show()

# Print the results
print("Comparison Indicators:")
print("MAE:", mae)
print("MSE:", mse)
print("MAPE:", mape)
print("RMSE:", rmse)
print("ARIMA Fitting Degree:", arima_fit_degree)
print("30-Day Predictions (after noise reduction):", average_predictions)
```
