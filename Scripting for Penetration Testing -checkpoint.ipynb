```python
import socket
import dpkt

# Function to extract and display information from a PCAP file
def process_pcap_file(file_path):
    with open(file_path, "rb") as pcap_file:
        pcap = dpkt.pcap.Reader(pcap_file)

        # Extract the IP addresses
        source_ip = ""
        target_ip = ""

        # Extract the total number of packets
        total_packets = 0

        # Extract the total number of ports
        total_ports = set()

        # Extract the scanned ports
        scanned_ports = set()

        for timestamp, packet_data in pcap:
            eth = dpkt.ethernet.Ethernet(packet_data)
            if isinstance(eth.data, dpkt.ip.IP):
                ip = eth.data
                if isinstance(ip.data, dpkt.tcp.TCP) or isinstance(ip.data, dpkt.udp.UDP):
                    if not source_ip:
                        source_ip = socket.inet_ntoa(ip.src)
                        target_ip = socket.inet_ntoa(ip.dst)

                    total_packets += 1

                    if isinstance(ip.data, dpkt.tcp.TCP):
                        total_ports.add(ip.data.dport)
                    elif isinstance(ip.data, dpkt.udp.UDP):
                        total_ports.add(ip.data.dport)

        # Display the extracted information
        print(f"IP performing the Nmap scan: {source_ip}")
        print(f"Targeted IP: {target_ip}")
        print(f"Total number of packets captured: {total_packets}")
        print(f"Total number of ports scanned: {len(total_ports)}")
        print("Scanned ports:")
        for port in sorted(total_ports):
            print(port)

        print("\n")
# Process each PCAP file
pcap_files = [
    r"C:\Users\This PC\Downloads\PCAP\ay2223_sasp_nmapcatured_1.pcap",
    r"C:\Users\This PC\Downloads\PCAP\ay2223_sasp_nmapcatured_2.pcap",
    r"C:\Users\This PC\Downloads\PCAP\ay2223_sasp_nmapcatured_3.pcap"
]

for pcap_file in pcap_files:
    print(f"Processing {pcap_file}:")
    process_pcap_file(pcap_file)
```
