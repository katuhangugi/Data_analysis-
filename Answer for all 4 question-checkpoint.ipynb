```python
import pandas as pd

# Load the dataset
dataset_path = r'C:\Users\This PC\Downloads\MKTG6018_final_assessment\question_2.csv'
df = pd.read_csv(dataset_path)

# Display the first few rows of the dataset
print(df.head())
```

```python
import scipy.stats as stats

# Perform pairwise t-tests between profile groups
profile_1_data = df[df['ProfileID'] == 1]['Seconds']
profile_2_data = df[df['ProfileID'] == 2]['Seconds']
profile_3_data = df[df['ProfileID'] == 3]['Seconds']

# Perform t-tests between the profile groups
t_statistic_1_vs_2, p_value_1_vs_2 = stats.ttest_ind(profile_1_data, profile_2_data)
t_statistic_1_vs_3, p_value_1_vs_3 = stats.ttest_ind(profile_1_data, profile_3_data)
t_statistic_2_vs_3, p_value_2_vs_3 = stats.ttest_ind(profile_2_data, profile_3_data)

# Print the t-statistics and p-values
print("Profile 1 vs. Profile 2: t-statistic =", t_statistic_1_vs_2, "p-value =", p_value_1_vs_2)
print("Profile 1 vs. Profile 3: t-statistic =", t_statistic_1_vs_3, "p-value =", p_value_1_vs_3)
print("Profile 2 vs. Profile 3: t-statistic =", t_statistic_2_vs_3, "p-value =", p_value_2_vs_3)
```

In light of the findings of the hypothesis tests, we are in a position to make inferences concerning the efficiency of the profiles with regard to the length of time spent engaged. We are able to conclude that there is a significant difference between the profiles if the p-value is lower than the significance level (0.01). This allows us to reject the null hypothesis and come to this conclusion.
For instance, if the p-value comparing Profile 1 and Profile 2 is less than 0.01, this suggests that there is a substantial difference in the amount of time spent engaged between these two profiles. We can discover which profiles are significantly distinct from one another by comparing the p-values of all of the paired tests and examining the results. We are able to establish which profile(s) demonstrate higher engagement duration by performing hypothesis testing, examining the data, and looking at the results.


# Question 3

```python
import pandas as pd

# Load the dataset
dataset_path = r'C:\Users\This PC\Downloads\MKTG6018_final_assessment\question_3.csv'
df = pd.read_csv(dataset_path)

# Display the first few rows of the dataset
print(df.head())
```

```python
# Check the data types and general information
print(df.info())

# Check basic statistical summary
print(df.describe())

# Check unique items in the "Item_Name" column
print(df['Item_Name'].unique())
```

```python
import matplotlib.pyplot as plt

# Calculate the total quantity ordered for each item
item_quantity = df.groupby('Item_Name')['Quantity'].sum()

# Sort the items in descending order of quantity
top_items = item_quantity.nlargest(5)

# Create a bar plot to visualize the most popular items
plt.figure(figsize=(10, 6))
top_items.plot(kind='bar')
plt.xlabel('Item Name')
plt.ylabel('Quantity Ordered')
plt.title('Top 5 Popular Food Items')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the total revenue generated by each item
item_revenue = df.groupby('Item_Name')['Total_products'].sum()

# Calculate the total cost of each item
item_cost = df.groupby('Item_Name')['Product_Price'].sum()

# Calculate the profitability of each item
item_profitability = item_revenue - item_cost

# Sort the items in descending order of profitability
top_profitable_items = item_profitability.nlargest(5)

# Create a bar plot to visualize the top profitable items
plt.figure(figsize=(10, 6))
top_profitable_items.plot(kind='bar')
plt.xlabel('Item Name')
plt.ylabel('Profitability')
plt.title('Top 5 Profitable Food Items')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the total number of orders for each customer
customer_order_count = df.groupby('Order_Number')['Item_Name'].count()

# Plot a histogram to visualize the distribution of order frequencies
plt.figure(figsize=(10, 6))
plt.hist(customer_order_count, bins=20)
plt.xlabel('Number of Orders')
plt.ylabel('Number of Customers')
plt.title('Distribution of Customer Order Frequencies')
plt.show()
```

```python
# Calculate the total order value for each customer
customer_order_value = df.groupby('Order_Number')['Total_products'].sum()

# Plot a histogram to visualize the distribution of order values
plt.figure(figsize=(10, 6))
plt.hist(customer_order_value, bins=20)
plt.xlabel('Order Value')
plt.ylabel('Number of Customers')
plt.title('Distribution of Customer Order Values')
plt.show()
```

We are able to gather insights into distinct client groups and create marketing strategies in accordance with those findings by segmenting customers based on the frequency of their orders, the value of their orders, and the items they prefer to purchase. On HungryPanda, the use of this segmentation can be helpful in the creation of targeted promotions, the improvement of customer targeting, and the enhancement of customer retention.

```python
import pandas as pd

# Load the dataset
dataset_path = r'C:\Users\This PC\Downloads\MKTG6018_final_assessment\question_4.xlsx'
df = pd.read_excel(dataset_path)

# Display the first few rows of the dataset
print(df.head())
```

```python
# Check the data types and general information
print(df.info())

# Check basic statistical summary of ratings
print(df['rating'].describe())

# Calculate the average rating
average_rating = df['rating'].mean()
print("Average Rating:", average_rating)
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download the required NLTK resources
nltk.download('vader_lexicon')

# Create an instance of the SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Define a function to extract sentiment scores from text
def get_sentiment_score(text):
    sentiment = sia.polarity_scores(text)
    return sentiment['compound']

# Apply the sentiment analysis function to review titles and contents
df['title_sentiment'] = df['review_title'].apply(get_sentiment_score)
df['content_sentiment'] = df['review_content'].apply(get_sentiment_score)

# Analyze the sentiments of the reviews
average_title_sentiment = df['title_sentiment'].mean()
average_content_sentiment = df['content_sentiment'].mean()
print("Average Title Sentiment:", average_title_sentiment)
print("Average Content Sentiment:", average_content_sentiment)
```

```python
import matplotlib.pyplot as plt

# Create a histogram of title sentiments
plt.figure(figsize=(10, 6))
plt.hist(df['title_sentiment'], bins=20)
plt.xlabel('Title Sentiment')
plt.ylabel('Frequency')
plt.title('Distribution of Title Sentiments')
plt.show()
```

```python
# Create a histogram of content sentiments
plt.figure(figsize=(10, 6))
plt.hist(df['content_sentiment'], bins=20)
plt.xlabel('Content Sentiment')
plt.ylabel('Frequency')
plt.title('Distribution of Content Sentiments')
plt.show()
```
