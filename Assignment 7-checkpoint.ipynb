```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
C_NTA_range = np.linspace(1e-6, 2e-4, 100)  # Concentration range of NTA (M)

# Equilibrium constants for complexation with NTA (example values, modify as needed)
K_CaNTA = 1e4
K_MgNTA = 1e5
K_ZnNTA = 1e6

# Empty lists to store the results
Ca2_plus = []
Mg2_plus = []
Zn2_plus = []

# Calculate the chemical speciation for each NTA concentration
for C_NTA in C_NTA_range:
    # Calculate the concentration of Ca2+
    Ca2_plus_concentration = C_Ca + K_CaNTA * C_NTA

    # Calculate the concentration of Mg2+
    Mg2_plus_concentration = C_Mg + K_MgNTA * C_NTA

    # Calculate the concentration of Zn2+
    Zn2_plus_concentration = C_Zn + K_ZnNTA * C_NTA

    # Append the results to the respective lists
    Ca2_plus.append(Ca2_plus_concentration)
    Mg2_plus.append(Mg2_plus_concentration)
    Zn2_plus.append(Zn2_plus_concentration)

# Plot the results
plt.plot(C_NTA_range, Ca2_plus, label='Ca2+')
plt.plot(C_NTA_range, Mg2_plus, label='Mg2+')
plt.plot(C_NTA_range, Zn2_plus, label='Zn2+')
plt.xlabel('NTA Concentration (M)')
plt.ylabel('Cation Concentration (M)')
plt.legend()
plt.show()
```
