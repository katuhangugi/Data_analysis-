```python
import csv 


with open("C:\\Users\\This PC\\Downloads\\enron (1).csv", newline='', encoding="utf8") as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        print(row)


#term search

def term_search(terms):
    terms_list = terms.split()
    matches = []
    with open('C:\\Users\\This PC\\Downloads\\enron (1).csv', newline='') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            body_list = row[2].lower().split()
            if all(term in body_list for term in terms_list):
                matches.append(row)
    print('Results found: {}'.format(len(matches)))
    for match in matches:
        print('{}. {} <{}> {}'.format(matches.index(match)+1, match[0], match[1], match[3]))


#address search

def address_search(name):
    last_name, first_name = name.split()
    matches = []
    with open('enron.csv', newline='') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            if row[0].lower() == last_name.lower() and row[1].lower() == first_name.lower():
                matches.append(row)
    print('Results found: {}'.format(len(matches)))
    for match in matches:
        print('{}. <{}>'.format(matches.index(match)+1, match[2]))


#interaction search 

def interaction_search(addresses):
    address_1, address_2 = addresses.split()
    matches = []
    with open('C:\\Users\\This PC\\Downloads\\enron (1).csv', newline='') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            if (row[2].lower() == address_1.lower() and row[3].lower() == address_2.lower()) or (row[2].lower() == address_2.lower() and row[3].lower() == address_1.lower()):
                matches.append(row)
    print('Results found: {}'.format(len(matches)))
    for match in matches:
        print('{}. <{}> -> <{}> [Subject: {}] {}'.format(matches.index(match)+1, match[2], match[3], match[4], match[5]))
```
