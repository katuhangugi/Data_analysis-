```python
import os

def analyze_directory(path):
    for root, directories, files in os.walk(path):
        for directory in directories:
            directory_path = os.path.join(root, directory)
            print(f"Directory: {directory_path}")

        for file in files:
            file_path = os.path.join(root, file)
            print(f"File: {file_path}")
            with open(file_path, 'r') as file_handle:
                print("File Contents:")
                line_number = 1
                for line in file_handle:
                    print(f"{line_number}: {line.strip()}")
                    line_number += 1
                print()

# Replace the path below with the directory you want to analyze
directory_path = r"C:\Users\This PC\Desktop\penetration"
analyze_directory(directory_path)
```

```python
import os
import matplotlib.pyplot as plt

folder_path = r'C:\Users\This PC\Desktop\penetation'

# Define keywords related to the application's code, configuration files, and network infrastructure
code_keywords = ['code', 'programming', 'function', 'method']
config_keywords = ['configuration', 'settings', 'parameters']
network_keywords = ['network', 'firewall', 'router']

# Initialize counters for each category
code_count = 0
config_count = 0
network_count = 0

# Get a list of files in the folder
files = os.listdir(folder_path)

# Iterate over each file and perform analysis
for file_name in files:
    file_path = os.path.join(folder_path, file_name)
    
    # Perform analysis on the file
    file_size = os.path.getsize(file_path)
    file_extension = os.path.splitext(file_name)[1]
    
    # Check if the file is a text file
    if file_extension == ".txt":
        with open(file_path, 'r') as file:
            file_content = file.read()
            
            # Count the occurrences of keywords in the file content
            code_count += sum(file_content.lower().count(keyword) for keyword in code_keywords)
            config_count += sum(file_content.lower().count(keyword) for keyword in config_keywords)
            network_count += sum(file_content.lower().count(keyword) for keyword in network_keywords)

# Prepare data for plotting
categories = ['Code', 'Configuration', 'Network']
counts = [code_count, config_count, network_count]

# Plot the graph
plt.bar(categories, counts)
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Occurrences of Keywords')
plt.show()
```
