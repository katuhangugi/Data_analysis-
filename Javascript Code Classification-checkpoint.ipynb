```python
import os

# Define folder paths
functional_folder = r"C:\Users\This PC\Desktop\functionalJS"
tracking_folder = r"C:\Users\This PC\Desktop\trackingJS"

# List of supported encodings to try
encodings = ['utf-8', 'latin-1', 'utf-16']

# Function to preprocess the code
def preprocess_code(code):
    # Add your preprocessing steps here
    return code

# Load functional JavaScript code samples
functional_samples = []
for filename in os.listdir(functional_folder):
    with open(os.path.join(functional_folder, filename), 'r', encoding='utf-8') as file:
        code = None
        for encoding in encodings:
            try:
                code = file.read()
                break
            except UnicodeDecodeError:
                continue
        if code is None:
            print(f"Unable to decode file: {filename}")
        else:
            functional_samples.append(preprocess_code(code))

# Load tracking JavaScript code samples
tracking_samples = []
for filename in os.listdir(tracking_folder):
    with open(os.path.join(tracking_folder, filename), 'r', encoding='utf-8') as file:
        code = None
        for encoding in encodings:
            try:
                code = file.read()
                break
            except UnicodeDecodeError:
                continue
        if code is None:
            print(f"Unable to decode file: {filename}")
        else:
            tracking_samples.append(preprocess_code(code))

# Print the number of functional and tracking samples
print(f"Number of functional samples: {len(functional_samples)}")
print(f"Number of tracking samples: {len(tracking_samples)}")
```

```python
# Preprocess code samples (remove comments and whitespaces)
# Implement your own code preprocessing logic here
# For example, using regular expressions

import re

def preprocess_code(code):
    # Remove single-line comments
    code = re.sub(r'//.*', '', code)
    
    # Remove multi-line comments
    code = re.sub(r'/\*(?:.|\n)*?\*/', '', code)
    
    # Remove extra whitespaces
    code = re.sub(r'\s+', ' ', code)
    
    return code

functional_samples = [preprocess_code(code) for code in functional_samples]
tracking_samples = [preprocess_code(code) for code in tracking_samples]
```

```python
# Combine functional and tracking samples
all_samples = functional_samples + tracking_samples

# Create labels for the samples (0 for functional, 1 for tracking)
labels = [0] * len(functional_samples) + [1] * len(tracking_samples)

# Initialize the TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Compute TF-IDF features
features = vectorizer.fit_transform(all_samples)
```

```python
# Split the dataset into train, validation, and test sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42)
```

```python
# Initialize the SVM classifier
svm = SVC()

# Train the SVM model
svm.fit(X_train, y_train)
```

```python
# Make predictions on the validation set
y_pred = svm.predict(X_val)

# Evaluate the model
report = classification_report(y_val, y_pred)
print(report)
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'C': [1, 10, 100],
    'gamma': [0.1, 0.01, 0.001],
    'kernel': ['rbf', 'linear']
}

# Perform grid search to find the best parameters
grid_search = GridSearchCV(svm, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Get the best parameters and retrain the model
best_params = grid_search.best_params_
svm = SVC(**best_params)
svm.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred_test = svm.predict(X_test)

# Evaluate the model on the test set
report_test = classification_report(y_test, y_pred_test)
print(report_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred_test)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='d', cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Compute the probabilities of the positive class
y_pred_prob = svm.decision_function(X_test)

# Compute the false positive rate and true positive rate
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)

# Compute the area under the ROC curve
roc_auc = roc_auc_score(y_test, y_pred_prob)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0, 1])
plt.ylim([0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np

# Get the weights (coefficients) of the SVM model
weights = svm.coef_.toarray().ravel()

# Sort the features based on their weights
indices = np.argsort(weights)

# Get the feature names (words) from the TF-IDF vectorizer
feature_names = vectorizer.get_feature_names()

# Print the top 10 most important features
print("Top 10 most important features:")
for i in range(1, 11):
    feature_index = indices[-i]
    feature_name = feature_names[feature_index]
    weight = weights[feature_index]
    print(f"{i}. {feature_name}: {weight}")
```

```python
import matplotlib.pyplot as plt

# Set the number of top features to visualize
top_n = 10

# Get the top N most important features and their weights
top_features = []
top_weights = []
for i in range(1, top_n+1):
    feature_index = indices[-i]
    feature_name = feature_names[feature_index]
    weight = weights[feature_index]
    top_features.append(feature_name)
    top_weights.append(weight)

# Plot the top N most important features
plt.figure(figsize=(10, 6))
plt.barh(range(top_n), top_weights, align='center')
plt.yticks(range(top_n), top_features)
plt.xlabel('Weight')
plt.ylabel('Feature')
plt.title('Top {} Most Important Features'.format(top_n))
plt.tight_layout()
plt.show()
```
