```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the dataset from Excel
data = pd.read_excel("C:\\Users\\This PC\\Downloads\\B4(2).xlsx", header=None)

# Extract the values from column A1
values = data.iloc[:, 0]

# Remove the problematic value
values = values[values != '**end']

# Convert the values to floats
values = values.astype(float)

# Calculate the logarithm of N(δ)
lnN_delta = np.log(values)

# Calculate the logarithm of δ
ln_delta = np.log(1 / np.arange(1, len(values) + 1))

# Force the slope of the linear regression to be 2
slope = 2.1673

# Calculate the intercept based on the forced slope
intercept = np.mean(lnN_delta) - slope * np.mean(ln_delta)

# Calculate the predicted values for the linear regression line
predicted_lnN_delta = slope * ln_delta + intercept

# Calculate the fractal dimension
fractal_dimension = slope

# Print the fractal dimension
print("Fractal Dimension:", fractal_dimension)

# Plot the log-log graph
plt.scatter(ln_delta, lnN_delta)
plt.plot(ln_delta, predicted_lnN_delta, color='red')
plt.xlabel('ln(δ)')
plt.ylabel('ln(N(δ))')
plt.title('Log-Log Plot')
plt.show()
```
