```python
import pandas as pd
import numpy as np

# Read the data from the Excel file
data = pd.read_excel('C:/Users/This PC/Downloads/B41.xlsx', header=None)
values = data.iloc[:, 0].values

# Convert values to numeric (handling non-numeric values)
values_numeric = pd.to_numeric(values, errors='coerce')

# Remove NaN values
values_numeric = values_numeric[~np.isnan(values_numeric)]

# Calculate the range and number of divisions for the box-counting method
min_value = np.min(values_numeric)
max_value = np.max(values_numeric)
range_values = max_value - min_value
num_divisions = 10  # You can adjust this value depending on the resolution you desire

# Initialize the counting array
count_array = np.zeros((num_divisions, num_divisions))

# Perform the box-counting method
for value in values_numeric:
    x = int((value - min_value) / range_values * num_divisions)
    if x == num_divisions:
        x -= 1
    count_array[x, x] += 1

# Calculate the fractal dimension
counts = np.sum(count_array > 0)
epsilon = range_values / num_divisions
fractal_dimension = np.log(counts) / np.log(1 / epsilon)

print("Fractal Dimension:", fractal_dimension)
```
