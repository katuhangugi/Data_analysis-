```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Step 1: Load the dataset
data = pd.read_csv("C:\\Users\\This PC\\Downloads\\gbpcny=x.csv")
exchange_rate = data['Close']

# Step 2: Basic Data Description
exchange_rate.plot(figsize=(10, 6))
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.title('Basic Trend of Exchange Rate')
plt.show()

# Step 3: ARIMA Modeling
# Perform differencing to make the series smooth
diff_exchange_rate = exchange_rate.diff().dropna()

# Perform ADF unit root test to check for stationarity/smoothness
from statsmodels.tsa.stattools import adfuller
result = adfuller(diff_exchange_rate)
p_value = result[1]

if p_value < 0.05:
    print("The series is already smooth/stationary.")
else:
    print("The series is not smooth/stationary and requires differencing.")

# Determine optimal ARIMA parameters using AIC criterion
model = ARIMA(diff_exchange_rate, order=(2, 0, 2))
model_fit = model.fit()

# Step 4: GWO-LSTM Prediction
# Normalize the residuals
scaler = MinMaxScaler(feature_range=(0, 1))
residuals = model_fit.resid.values.reshape(-1, 1)
scaled_residuals = scaler.fit_transform(residuals)

# Split data into train and test sets
train_size = int(len(scaled_residuals) * 0.8)
train, test = scaled_residuals[:train_size], scaled_residuals[train_size:]

# Create sequences for LSTM training
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 1
X_train, y_train = create_sequences(train, seq_length)
X_test, y_test = create_sequences(test, seq_length)

# Build and train LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(units=50, activation='relu', input_shape=(seq_length, 1)))
model_lstm.add(Dense(units=1))
model_lstm.compile(optimizer='adam', loss='mean_squared_error')
model_lstm.fit(X_train, y_train, epochs=10, batch_size=1, verbose=2)

# Make predictions
train_predictions = model_lstm.predict(X_train)
test_predictions = model_lstm.predict(X_test)

# Invert predictions to the original scale
train_predictions = scaler.inverse_transform(train_predictions)
test_predictions = scaler.inverse_transform(test_predictions)

# Step 6: Evaluation
def evaluate_forecast(actual, predicted):
    mse = mean_squared_error(actual, predicted)
    mae = mean_absolute_error(actual, predicted)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((actual - predicted) / actual)) * 100
    return mse, mae, rmse, mape

train_actual = residuals[seq_length:train_size]
test_actual = residuals[train_size+seq_length:]

mse_train, mae_train, rmse_train, mape_train = evaluate_forecast(train_actual, train_predictions)
mse_test, mae_test, rmse_test, mape_test = evaluate_forecast(test_actual, test_predictions)

print("Training set:")
print("MSE:", mse_train)
print("MAE:", mae_train)
print("RMSE:", rmse_train)
print("MAPE:", mape_train)

print("\nTest set:")
print("MSE:", mse_test)
print("MAE:", mae_test)
print("RMSE:", rmse_test)
print("MAPE:", mape_test)

# Plot predicted values vs actual values
plt.figure(figsize=(10, 6))
plt.plot(train_actual, label='Actual (Training)')
plt.plot(train_predictions, label='Predicted (Training)')
plt.plot(test_actual, label='Actual (Test)')
plt.plot(test_predictions, label='Predicted (Test)')
plt.xlabel('Time')
plt.ylabel('Residuals')
plt.title('Predicted vs Actual Residuals')
plt.legend()
plt.show()


```

```python
import numpy as np
from pyhht import EMD
import matplotlib.pyplot as plt
import pywt

# Step 1: Define or load the original series in the variable 'exchange_rate'
exchange_rate = data['Close'] 

# Step 2: Decompose the original series using CEEMDAN
emd = EMD(exchange_rate)
imfs = emd.emd()

# Step 3: Add positive and negative Gaussian white noise to the original series
std_dev = 0.2
num_noise_additions = 100
noise_added_series = np.zeros_like(exchange_rate)

for _ in range(num_noise_additions):
    noise = np.random.normal(0, std_dev, size=len(exchange_rate))
    noise_added_series += noise

# Step 4: Perform CEEMDAN decomposition on the noise-added series
emd_noise_added = EMD()
imfs_noise_added = emd_noise_added.emd(noise_added_series, n_imfs=10)

# Step 5: Calculate the Pearson correlation coefficient with the original series
correlations = []
for component in imfs_noise_added:
    correlation = np.corrcoef(exchange_rate, component)[0, 1]
    correlations.append(correlation)

# Step 6: Apply wavelet packet noise reduction for correlations less than 0.2
threshold = 0.2
reconstructed_series = np.zeros_like(exchange_rate)

for i, component in enumerate(imfs_noise_added):
    if correlations[i] < threshold:
        # Apply wavelet packet noise reduction
        wp = pywt.WaveletPacket(component, 'db2', maxlevel=3)
        wp.reconstruct(update=True)
        reconstructed_component = wp.data
    else:
        reconstructed_component = component

    # Add the reconstructed component to the final decomposed data sequence
    reconstructed_series += reconstructed_component

# Plot the original series, decomposed components, and reconstructed series
plt.figure(figsize=(10, 6))
plt.plot(exchange_rate, label='Exchange Rate')
for i, component in enumerate(imfs):
    plt.plot(component, label=f'Component {i+1}')
plt.plot(reconstructed_series, label='Reconstructed Series')
plt.xlabel('Time')
plt.ylabel('Exchange Rate')
plt.title('CEEMDAN Decomposition and Noise Reduction')
plt.legend()
plt.show()
```
