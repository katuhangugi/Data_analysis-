````python
# Molar masses in g/mol
molar_masses = {
    'Ca++': 40.08,
    'Mg++': 24.31,
    'Na+': 22.99,
    'K+': 39.10,
    'HCO3-': 61.02,
    'SO4=': 96.06,
    'Cl-': 35.45,
    'SiO2': 60.09
}

# Valences
valences = {
    'Ca++': 2,
    'Mg++': 2,
    'Na+': 1,
    'K+': 1,
    'HCO3-': 1,
    'SO4=': 2,
    'Cl-': 1,
    'SiO2': 0
}

# Concentrations in mg/L
concentrations_mg_per_L = {
    'Ca++': 72.8,
    'Mg++': 38.8,
    'Na+': 51.4,
    'K+': 6.5,
    'HCO3-': 266.5,
    'SO4=': 120.0,
    'Cl-': 65.1,
    'SiO2': 12.1
}

# Conversion factors
mg_to_mmol = {ion: 1 / molar_masses[ion] for ion in molar_masses}
mmol_to_meq = {ion: valences[ion] for ion in valences}

# Calculate concentrations in mM
concentrations_mmol_per_L = {ion: concentrations_mg_per_L[ion] * mg_to_mmol[ion] for ion in concentrations_mg_per_L}

# Calculate concentrations in meq/L
concentrations_meq_per_L = {ion: concentrations_mmol_per_L[ion] * mmol_to_meq[ion] for ion in concentrations_mmol_per_L}

# Print results
print("Concentrations in mM:")
for ion, concentration in concentrations_mmol_per_L.items():
    print(ion, concentration)

print("\nConcentrations in meq/L:")
for ion, concentration in concentrations_meq_per_L.items():
    print(ion, concentration)
````

````python
# Calculate total positive charge (cations)
total_positive_charge = sum(concentrations_meq_per_L[ion] for ion in concentrations_meq_per_L if valences[ion] > 0)

# Calculate total negative charge (anions)
total_negative_charge = sum(concentrations_meq_per_L[ion] for ion in concentrations_meq_per_L if valences[ion] < 0)

# Check if solution is electroneutral
is_electroneutral = total_positive_charge == total_negative_charge

# Print results
print("Total positive charge (cations):", total_positive_charge, "meq/L")
print("Total negative charge (anions):", total_negative_charge, "meq/L")
print("Is the solution electroneutral?", is_electroneutral)
````

````python
# Calculate total positive charge (cations)
total_positive_charge = sum(concentrations_meq_per_L[ion] for ion in concentrations_meq_per_L if valences[ion] > 0)

# Calculate total negative charge (anions)
total_negative_charge = sum(concentrations_meq_per_L[ion] for ion in concentrations_meq_per_L if valences[ion] < 0)

# Calculate the ionic balance percentage
ionic_balance_percentage = abs(total_positive_charge - total_negative_charge) / (total_positive_charge + total_negative_charge) * 100

# Determine if there is an excess of anions or cations
if total_positive_charge > total_negative_charge:
    excess_ions = "cations"
    excess_percentage = (total_positive_charge - total_negative_charge) / (total_positive_charge + total_negative_charge) * 100
elif total_positive_charge < total_negative_charge:
    excess_ions = "anions"
    excess_percentage = (total_negative_charge - total_positive_charge) / (total_positive_charge + total_negative_charge) * 100
else:
    excess_ions = None
    excess_percentage = 0

# Print results
print("Total positive charge (cations):", total_positive_charge, "meq/L")
print("Total negative charge (anions):", total_negative_charge, "meq/L")

if excess_ions:
    print("There is an excess of", excess_ions, "by", excess_percentage, "%")
else:
    print("The solution is electroneutral.")

print("Ionic balance percentage:", ionic_balance_percentage, "%")
````
